```python
# Exercise - Recursion:
#Task 1 - Sum of Nested Lists:
# Function to calculate the sum of a nested list
def sum_nested_list(nested_list):
   
    total = 0
    for element in nested_list:
        if isinstance(element, list):  
            total += sum_nested_list(element)
        else:
            total += element 
    return total
    
# A nested list with multiple levels of nesting
nested_list_1 = [1, [2, [3, 4], 5], 6, [7, 8]]

# Call the function and print the result
print("Sum of nested_list_1:", sum_nested_list(nested_list_1))  
```

```python
#task2
# Function to generate all unique permutations of a string
def generate_permutations(s):
    if len(s) == 1:
        return [s]
 # Use a set to store permutations to avoid duplicates
    permutations = set()
    for i, char in enumerate(s):
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            permutations.add(char + perm)

    return list(permutations)
# Generate and print sorted permutations for the string "abc"
print(sorted(generate_permutations("abc")))
# Generate and print sorted permutations for the string "aab" 
print(sorted(generate_permutations("aab")))
```

```python
#Task 3 - Directory Size Calculation:

# Function to calculate the total size of a directory structure 
def calculate_directory_size(directory):
   
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict): 
            total_size += calculate_directory_size(value) 
        else:
            total_size += value  # If it's an integer, it's a file size
    return total_size
    
# Directory structure with files and subdirectories
directory_structure = {
    "file1.txt": 500,
    "file2.txt": 700,
    "subdir1": {
        "file3.txt": 200,
        "file4.txt": 300
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 150
        },
        "file6.txt": 550
    }
}

# Calculate and print the total size of the directory structure
total_size = calculate_directory_size(directory_structure)
print("Total Directory Size:", total_size, "KB")
```

```python
#Exercises - Dynamic Programming:
#Task 1 - Coin Change Problem:

# Function to find the minimum number of coins required to make a given amount
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Example
coins = [1, 2, 5]
amount = 11
print(min_coins(coins, amount))

```

```python
#task2

# Function to compute the length of the Longest Common Subsequence 
def longest_common_subsequence(s1, s2):
    dp = [[0] * (len(s2) + 1) for _ in range(len(s1) + 1)]
# Fill the DP table by iterating over each character of both strings
    for i in range(1, len(s1) + 1):
        for j in range(1, len(s2) + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    return dp[len(s1)][len(s2)]

s1 = "abcde"
s2 = "ace"
print(longest_common_subsequence(s1, s2))
```

```python
#Task3

# Function to calculate the maximum value that can be obtained with a given weight capacity
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
 # Iterate over each item (i) and each possible weight capacity (w)
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]
    # The maximum value we can achieve with the full set of items and the given capacity is found at dp[n][capacity]
    return dp[n][capacity]

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
# Print the result
print(knapsack(weights, values, capacity))
```
